```python
import requests

import bs4
from bs4 import BeautifulSoup
```

```python
start_url = 'https://www.gutenberg.org/ebooks/author/783'
base_url = 'https://www.gutenberg.org'

def get_text(uid, title):
    uid = str(uid)
    target_url = base_url + '/cache/epub/' + uid + '/pg' + uid + '.txt'
    print (target_url)
    result = requests.get(target_url)
    title = (title.replace(' ', '_')) + '.txt'
    if result.status_code == 200:
        file = open('texts/' + title, 'w')
        file.write(result.text)
        file.close

        
def get_links(soup):
    for li in soup.find_all('li', class_='booklink'):
        title_li= li.find('span', class_='title')
        title = title_li.text
        link = li.find('a', class_='link')
        href = link.get('href')
        uid = href.split("/")[2]
        print (uid)
        get_text(uid, title)

        
def get_pages(url):
    result = requests.get(url)
    if (result.status_code == 200):
        soup = BeautifulSoup(result.text)
        get_links(soup)
    next = soup.find(title='Go to the next page of results.')
    if next != None:
        href = next.get('href')
        print(href)
        new_url = base_url + (next.get('href'))
        print(new_url)
        get_pages(new_url)

        
get_pages(start_url)
```
